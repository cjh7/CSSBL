```python
os.chdir('/content/drive/My Drive/Colab Notebooks/CSSBL/data/real_kaveh/AUC')
result = pd.read_csv('propose_kaveh_real_AUC.csv',header=None)
propose_sim_AUC=result.values

os.chdir('/content/drive/My Drive/Colab Notebooks/CSSBL/data/real_kaveh/AUC')
result = pd.read_csv('lee_kaveh_real_AUC.csv',header=None)
LEE_sim_AUC=result.values

os.chdir('/content/drive/My Drive/Colab Notebooks/CSSBL/data/real_kaveh/AUC')
result = pd.read_csv('MSBL_kaveh_real_AUC.csv',header=None)
MSBL_sim_AUC=result.values

os.chdir('/content/drive/My Drive/Colab Notebooks/CSSBL/data/real_kaveh/AUC')
result = pd.read_csv('Dai_kaveh_real_AUC.csv',header=None)
DAI_sim_AUC=result.values

os.chdir('/content/drive/My Drive/Colab Notebooks/CSSBL/data/real_kaveh/AUC')
result = pd.read_csv('KAVEH_kaveh_real_AUC.csv',header=None)
KAVEH_sim_AUC=result.values
```

```python


for i in range(5):
  print(i)
  stat, p = ttest_rel(propose_sim_AUC[i,:] , KAVEH_sim_AUC[i,:], axis=0, alternative='greater')
  # stats.ttest_rel(P1, P2, axis=0, alternative='less')
  print('stat=%.3f, p=%.3f' % (stat, p))
  if p > 0.05:
    print('Probably the same distribution')
  else:
    print('Probably different distributions')
```
